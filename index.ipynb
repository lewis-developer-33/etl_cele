## Extract dataset

```python
import pandas as pd

df_customers = pd.read_csv('input/olist_customers_dataset.csv')
df_orders = pd.read_csv('input/olist_orders_dataset.csv')
df_sellers = pd.read_csv('input/olist_sellers_dataset.csv')
df_products = pd.read_csv('input/olist_products_dataset.csv')
df_payments = pd.read_csv('input/olist_order_payments_dataset.csv')
```

## Load

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///etl.db', echo=False)
```

```python
df_orders.to_sql(name='orders',con=engine,index=False,if_exists='replace')
df_customers.to_sql(name='customers',con=engine,index=False,if_exists='replace')
df_payments.to_sql(name='payments',con=engine,index=False,if_exists='replace')
df_products.to_sql(name='products',con=engine,index=False,if_exists='replace')
df_sellers.to_sql(name='sellers',con=engine,index=False,if_exists='replace')
```
